```python
from matplotlib.pyplot import figure
figure(figsize=(12, 6), dpi=80)
plt.plot(df.groupby(pd.Grouper(key="Transaction date",freq="14D")).count()['Basket size (units)'],'.-')
plt.xlabel("Date")
```

```python
fig, ax1 = plt.subplots()
fig.set_size_inches(18.5, 10.5)
ax2 = ax1.twinx()
ax1.plot( df.groupby(pd.Grouper(key="Transaction date",freq="14D")).count()['Customer ID'], 'g-',c='red')
ax2.plot( df.groupby(pd.Grouper(key="Transaction date",freq="14D")).mean()['Basket size (units)'], 'g-');
```

Hypothesis test

```python
from scipy import stats
```

```python
before=df.loc[df["Transaction date"]<="2018-05-20"]['Basket size (dollars)']
```

```python
after=df.loc[df["Transaction date"]>="2018-05-20"]['Basket size (dollars)']
```

```python
stats.ttest_ind(before, after, equal_var=False)
```
